```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.random(100)
y = np.random.random (100)

classes = np.random.randint (0, 3, 100)
plt.scatter (x, y, c = classes)
plt.show()
```

```python
# List of colormaps to use
colormaps = ['cool', 'Set1', 'Set2', 'Spectral', 'Purples', 'Greys']

# Create a figure for subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()  # Flatten the 2D array of axes

# Create scatter plots for each colormap
for i, cmap in enumerate(colormaps):
    scatter = axs[i].scatter(x, y, c=classes, cmap=cmap)
    axs[i].set_title(f'Scatter Plot with {cmap} Colormap')
    axs[i].set_xlabel('X-axis')
    axs[i].set_ylabel('Y-axis')
    plt.colorbar(scatter, ax=axs[i], label='Classes')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
