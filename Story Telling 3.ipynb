```python
#import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
import plotly.express as px
```

```python
# Creating a sample dataset
np.random.seed(42)
data = {
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 500),
    'Product_Category': np.random.choice(['Electronics', 'Clothing', 'Home Goods'], 500),
    'Sales': np.random.normal(200, 50, 500),
    'Profit': np.random.normal(50, 20, 500),
    'Customer_Age': np.random.randint(18, 70, 500),
    'Customer_Segment': np.random.choice(['Young Adult', 'Adult', 'Senior'], 500),
    'Date': pd.date_range(start='2023-01-01', periods=500, freq='D')
}

df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
# Principle of Preattentive Attributes - Color and Size
#Preattentive attributes like color and size can be processed by the brain almost instantly. 
#Let's use these attributes to highlight key insights.
# Using color to highlight high-profit regions in a scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Sales'], df['Profit'], c=df['Profit'], cmap='RdYlGn', s=50)
plt.colorbar(scatter, label='Profit')
plt.title('Sales vs Profit with Profit Highlighted by Color')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

This scatter plot uses color to instantly draw attention to regions with higher profits. 
The use of the RdYlGn colormap (red-yellow-green) emphasizes areas of higher profit (green), 
helping viewers focus on the most profitable areas at a glance.

```python
#Gestalt Principle - Proximity and Similarity
#The Gestalt principles explain how people tend to group visual elements.
#We'll use proximity and similarity to tell a story about customer segments.
# Boxplot to show the distribution of customer age by segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Customer_Segment', y='Customer_Age', data=df, palette='pastel')
plt.title('Customer Age Distribution by Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Customer Age')
plt.show()
```

In this boxplot, the Gestalt principles of proximity and similarity are used. 
By grouping ages within segments, its easy to compare age distributions across different customer segments. 
The similarity in color helps reinforce the grouping, making the comparison intuitive.

```python
#Principle of Figure and Ground - Highlighting Important Information
#Using figure-ground contrast, you can highlight important parts of your data by making them stand out from the background.
# Highlighting a specific region with a different color
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=df, palette=['grey' if x != 'North' else 'red' for x in df['Region']])
plt.title('Sales by Region with North Region Highlighted')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

In this bar plot, the 'North' region is highlighted using a different color (red), making it stand out against 
the grey background of other regions. This figure-ground distinction ensures that the 'North' region captures 
immediate attention.

```python
#Principle of Enclosure - Grouping Related Data
#Enclosure can be used to visually group related data points. This can help in organizing information and guiding the 
#viewers focus.
# Heatmap to show correlation between numerical variables with a focus on Sales
corr = df[['Sales', 'Profit', 'Customer_Age']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=2, linecolor='black')
plt.title('Correlation Between Sales, Profit, and Customer Age')
plt.show()
```

The heatmap uses color and enclosure (grid lines) to highlight correlations between variables. 
The use of lines and contrasting colors helps in distinguishing different sections of the heatmap, 
making it easier to focus on areas of interest.

```python
#Using Multiple Perception Principles in an Interactive Plot
#Combining multiple perception principles, like color, proximity, and size, we can create a more complex visualization. 
# Interactive scatter plot using Plotly
fig = px.scatter(df, x='Sales', y='Profit', color='Product_Category', size='Customer_Age', 
                 hover_data=['Region', 'Customer_Segment'], title='Sales vs Profit by Product Category and Age')
fig.update_traces(marker=dict(opacity=0.6, line=dict(width=2, color='DarkSlateGrey')))
fig.show()
```

This interactive scatter plot uses color to differentiate product categories, size to represent customer age, 
and hover information to provide additional context. By combining these visual variables, it creates a rich and 
intuitive visualization that is easy to explore.

This notebook demonstrates how to use perception principles in data visualization to create more effective and 
compelling stories. By leveraging preattentive attributes, Gestalt principles, figure-ground contrast, and 
enclosure, you can guide your audiences attention, emphasize key insights, and make your data story more memorable.
